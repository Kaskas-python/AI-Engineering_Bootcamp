### Import Dependencies

```python
from langsmith import Client
```

#### Attach Feedback to a specific trace in LangSmith

019c3d9c-85cf-74a0-92f7-2686a7276d37

```python
client = Client()
```

```python
RUN_ID = "019c3d9c-85cf-74a0-92f7-2686a7276d37"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments",
    value="This is a long form free text feedback left by Arminas",
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments_with_score",
    value="This is a long form free text feedback left by Arminas",
    score=0,
    feedback_source_type="api"
)
```
